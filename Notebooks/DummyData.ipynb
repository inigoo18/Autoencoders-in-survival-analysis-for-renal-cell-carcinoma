```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Number of samples and features
num_samples = 1000
num_features = 5000

# Generate random data from a normal distribution
data = np.random.normal(loc=0.5, scale=0.15, size=(num_samples//2, num_features))
data = np.hstack((data, np.ones((data.shape[0], 1))))
data2 = np.random.normal(loc=1, scale=0.15, size=(num_samples//2, num_features))
data2 = np.hstack((data2, 2* np.ones((data2.shape[0], 1))))

final_data = np.concatenate((data, data2), axis=0)
print(final_data.shape)

column_names = [f'F{i}' for i in range(num_features+1)]
index_names = [f'S{i}' for i in range(num_samples)]

# Create DataFrame
df = pd.DataFrame(final_data, columns = column_names, index = index_names)

# Display the DataFrame
df
```

```python
df.to_csv('dummy_data.csv')
```
