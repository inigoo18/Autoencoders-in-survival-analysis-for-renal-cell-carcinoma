```python
ad_video_ids = []

for ad_idx in list(df_ad.index):
    ad = df_ad.iloc[ad_idx]
    ad_video_ids += [[ad['long_version_video_id'], ad['short_version_video_id']]]

ad_video_ids
```

```python
cnt = 0
for ad_vid in ad_video_ids:
    if df_vid.iloc[ad_vid[0]]['duration'] < df_vid.iloc[ad_vid[1]]['duration']:
        cnt += 1
cnt
```

```python
lastStr = ""

for i in range(len(df)):
    pat = df.loc[i]
    actStr = pat['date']
    if actStr != '-':
        lastStr = actStr
    else:
        pat['date'] = lastStr
        
    df.loc[i] = pat
```
